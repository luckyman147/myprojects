```python
import pandas as pd  

df=pd.read_csv('school.csv')
r=df.corr()['height']['weight']
print("correlation between height and weight is {:.2f}".format(r))
```

```python
df1=pd.read_csv('auto-mpg.csv')
r=df1.corr()['mpg']['weight']
print("correlation between height and weight is {:.2f}".format(r))
```

Positive correlation ( 0<r≤1 ) means that as one variable increases, the other also increases. They move in the same direction.

Negative correlation ( −1≤r<0 ) means that as one variable increases, the other decreases. They move in opposite directions.

A zero correlation ( r=0 ) means that the two variables are totally independent from each other.

```python
import numpy as np
import matplotlib.pyplot as plt 

dfC=pd.read_csv('anescombes.csv',sep=';')
dfC.head()
```

```python
print( "mean y group:\n",dfC.groupby(by='dataset').mean())
print("std by group :\n",dfC.groupby(by='dataset').std())
```

```python
fig,ax=plt.subplots(2,2,figsize=(6,6))

for k,name in np.ndenumerate(dfC.dataset.unique()):
    cond=dfC.dataset==name
    sing=np.polyfit(dfC[cond].x,dfC[cond].y,1)
    sing_x=np.linspace(dfC[cond].x.min(),dfC[cond].x.max(),3)
    sing_y=sing[0]*sing_x+sing [1]
    plt.subplot(2,2,k[0]+1)
    plt.scatter(dfC[cond].x,dfC[cond].y)
    plt.plot(sing_x,sing_y,color='orange')
```

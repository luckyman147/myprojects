```python
import pandas as pd
from sklearn import datasets,model_selection,preprocessing,tree,metrics,ensemble,svm
```

import data


```python
iris_data=datasets.load_iris()
x,y=datasets.load_iris(return_X_y=True)
x_normalize=preprocessing.StandardScaler()
x_norm=x_normalize.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(x_norm,y,test_size=0.1,random_state=42,stratify=y)
cl=tree.DecisionTreeClassifier()
cl.fit(x_train,y_train)
y_pred=cl.predict(x_test)
metrics.accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(10,10),dpi=180)

tree.plot_tree(cl)
plt.savefig('tree2.png')
```

```python
x,y=datasets.load_breast_cancer(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.1,random_state=42,stratify=y)
```

```python
rf=ensemble.RandomForestClassifier(random_state=42,n_estimators=100,max_depth=5,n_jobs=-1)
rf.fit(x_train,y_train)
y_predict=rf.predict(x_test)
metrics.accuracy_score(y_test,y_predict)
```

suppert vector classifier

```python
clf=svm.SVC(kernel='linear')
clf.fit(x_train,y_train)
y_pre=clf.predict(x_test)
metrics.accuracy_score(y_test,y_pre)
```

```python
cm=metrics.confusion_matrix(y_test,y_pre)
tp,tn,fp,fn=cm[1,1],cm[0,0],cm[0,1],cm[1,0]
print(f'True Positive :{tp}')
print(f'True negative :{tn}')
print(f'false Positive :{fp}')
print(f'false negative :{fn}')
print("precison",metrics.precision_score(y_test,y_pre))
print("recall",metrics.recall_score(y_test,y_pre))
print("f1_score",metrics.f1_score(y_test,y_pre))
metrics.plot_confusion_matrix(clf,x_test,y_test)
```

```python
k,l=datasets.fetch_california_housing(return_X_y=True)
l_mean=l.mean()
l_std=l.std()
l_normalize=(l-l_mean)/l_std
k_mean=k.mean()
k_std=k.std()
k_normalize=(k-k_mean)/k_std
```

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(k_normalize,l_normalize,test_size=0.2,random_state=42)

clR=svm.SVR()
clR.fit(x_train,y_train)
y_pre=clR.predict(x_test)

```

```python
metrics.r2_score(y_test,y_pre),metrics.mean_absolute_error(y_test,y_pre)

```

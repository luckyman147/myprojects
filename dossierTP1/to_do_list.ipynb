```python
#pip install openpyxl
import openpyxl
from openpyxl.styles import PatternFill
# Create a new Excel file and add a sheet
wb = openpyxl.Workbook()
sheet = wb.active
sheet.title = "To-Do List"

# Add some data to the sheet
sheet["A1"] = "Monday"
sheet["B1"] = "tuesday"
sheet["C1"] = "wednesday"
sheet["D1"] = "thursday"
sheet["E1"] = "friday"
sheet["F1"] = "Saturday"
sheet["G1"] = "Sunday"
sheet["H1"] = "importance"


# Modify the data
for i in range(2,50):
    for k in range(65,91):
        sheet[f"{chr(k)}{i}"] = ""
    

# Save the file
wb.save("todo_lt.xlsx")
```

```python
wb = openpyxl.load_workbook("todo_list.xlsx")

fill_color = "00FF00"  # Green color

# Create a new fill style
fill = PatternFill(start_color=fill_color, end_color=fill_color, fill_type="solid")

# Loop through all cells in the column and add the fill color
for row in sheet.iter_rows(min_row=1, min_col=1, max_col=1):
    for cell in row:
        cell.fill = fill
wb.save("todo_list.xlsx")

```
